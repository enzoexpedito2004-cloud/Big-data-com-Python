```python
#importando a biblioteca pandas para analise e numpy para controlar o volume de dados
import pandas as pd
import numpy as np
# Fiz uma variavel "vendas" recebendo a planilha para nos analisar 
vendas=pd.read_excel("BaseFuncionarios.xlsx")
display(vendas)
```

```python
#Tipos de inspeção que eu aprendi
# sample mostra uma linha aleatoria
display(vendas.sample())
#shape
display(vendas.shape)
display(vendas.columns)
display(vendas.info())
display(vendas.describe())
```

```python
#limpesa e tratamento
#display(vendas["Nome Cliente"])
lista=["Nome Cliente","Produto"]
#display(vendas[lista])
exclua=vendas.drop(columns=["Horas Extras"])
#display(exclua)
#display(vendas)
exclua["Data de Demissao"]=exclua["Data de Demissao"].fillna("vazio")
exclua["Data de Contratacao"]=pd.to_datetime(exclua["Data de Contratacao"],format="%Y-%m-%d")
display(exclua)
display(exclua.info())
exclua["Nome"]=exclua["Nome"].str.strip()
exclua["Nome"]=exclua["Nome"].str.title()
exclua=exclua.drop_duplicates(subset="ID RH")
display(exclua)
```

```python
exclua=vendas
exclua["media avalia"]=exclua["Avaliação do Funcionário"].sum()/exclua["Nome"].count()
exclua["resumo"]=np.where(exclua["Avaliação do Funcionário"]>8,"satisfeito","insatisfeito")
Regioes_estado={"Rio de Janeiro":"Suldeste","São Paulo":"Suldeste","Belo Horizonte":"Suldeste","Campinas":"Suldeste",
'Salvador':"Nordeste","Recife":"nordeste",
"Curitiba":"Sul","Porto Alegre":"Sul"
,"Goiânia":"Centro Oeste"}
exclua["Região"]=exclua["Cidade"].map(Regioes_estado)
display(exclua["Cidade"].unique())
display(exclua)
```
